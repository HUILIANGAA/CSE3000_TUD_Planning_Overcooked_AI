```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# def showHist(source_list, title, ylabel):

#     configurations, avg_reward = zip(*source_list)

#     x = np.arange(len(source_list_cramped))
#     plt.bar(x, height= avg_reward)
#     plt.xticks(x, configurations, rotation=90)
#     plt.title(title)
#     plt.ylabel(ylabel)
#     plt.show()
    
def showHist_color(source_list, title, ylabel, save, color=["b", "g", "r", "c", "m", "y", "k", "w"]):

    configs = []
    rewards = []
    rewards_app = []
    reward_indices = []
    
    for source_l in source_list:
        configurations, reward = zip(*source_l)
        configs.extend(configurations)
        rewards.extend(reward)
        rewards_app.append(reward)
    
    for r in rewards_app:
        reward_indices.append(len(r))
    
#     print("configs", configs)
#     print("rewards", rewards)
#     print("r2: ", rewards_app)
#     print("indices", reward_indices)
    
    x = np.arange(sum(reward_indices))
    barlist = plt.bar(x, height= rewards)
    
    counter = 0
    for c_i, i in enumerate(reward_indices):
        for j in range(counter, counter+i):
#             print(i,c_i,counter)
            barlist[j].set_color(color[c_i])
            
            if configs[j] == "CP+H$_{Proxy}$-400-5R" and "Cramped" in title and "Performance" in title:
                plt.errorbar(j, rewards[j], yerr=32.20, color="black")
    
        counter += i
    plt.plot(0,0, label="CP+CP", color="b")
    plt.plot(0,0, label="CP$_{X}$+CP$_{X}$", color="g")
    plt.plot(0,0, label="CP+H$_{Proxy}$", color="r")
    plt.plot(0,0, label="CP$_{X}$+H$_{Proxy}$", color="c")
    plt.plot(0,0, label="P$_{H_{Proxy}}$+H$_{Proxy}$", color="m")
    plt.plot(0,0, label="P$_{BC}$+H$_{Proxy}$", color="y")
    
    plt.rcParams["figure.figsize"] = (10,3)
    plt.legend(loc="best", ncol=2)
    plt.xticks(x, configs, rotation=90)
    plt.title(title)
    plt.ylabel(ylabel)
    
    plt.savefig(save + ".png", format='png', bbox_inches='tight')
    
    plt.show()
```
